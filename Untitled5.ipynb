```python
def plot_emotions(dataframe, x=WINDOW_LENGTH, summarize=True):
    #dataframe.drop('file_name', 1, inplace=True)
    html = ''
    for face in dataframe['face#'].unique():
        data = dataframe[dataframe['face#']==face].copy(deep=True)
        data.drop('face#', 1, inplace=True)
        data.set_index('frame#', inplace=True)
        if summarize==True:
            d2 = DataFrame.rolling(data, x).mean()
            #d3 = DataFrame.rolling(data, x).std()
            #d2.columns = [str(col) + '_mean'+'(x=%d)'%x for col in d2.columns ]
            d3 = DataFrame.rolling(data, x).std()
            std = pandas.concat([d3[col].sort_values(ascending=False)\
                     .reset_index(drop=True) for col in d3], axis=1, ignore_index=True)
            std1=std.head(3)
            std1_min = std1.loc[:].min().tolist()
            d3[(d3.iloc[:] >  min (std1_min))]=0.5
            d3[(d3.iloc[:] < min (std1_min))]= np.nan
            html += line_plot(data, title='face num:-%s'%face)
            html += '<br> <br>'

        else:
            html += line_plot(data[EMOTION_LABELS],
                                title='face num:-%s'%face)
            html += '<br> <br>'
    return html

def line_plot(dataframe, title='face0', **kwargs):
    dataframe.sort_index(inplace=True)
    mypalette = np.random.choice(Magma[10], size=len(dataframe.columns), replace=False)
    p = figure(width=1000, height=600, x_axis_type='datetime', title=title)
    xs = dataframe.index.values
    d3 = DataFrame.rolling(data, x).std()
    std = pandas.concat([d3[col].sort_values(ascending=False)\
                     .reset_index(drop=True) for col in d3], axis=1, ignore_index=True)
    std1=std.head(3)
    std1_min = std1.loc[:].min().tolist()
    d3[(d3.iloc[:] >  min (std1_min))]=0.5
    d3[(d3.iloc[:] < min (std1_min))]= np.nan
    xs1=d3.index.values
    for (color,label) in zip(mypalette, dataframe.columns):
        p.line(xs, dataframe[label], color=color, legend=label)
        p.rect(xs1,d3[label],width=1, height=1200, width_units="screen",\
               height_units="screen")
    html = file_html(p, CDN, 'emotions plot')
    return html
```
